```python
import numpy as np
import random
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
def randomwalk1D(n):
    x = 0
    start = x
    xposition = [start]
    probabilities = [-1, 1]
    for i in range(1, n + 1):
        x += random.choice(probabilities)
        xposition.append(x)
    return xposition
```

```python
plt.plot(randomwalk1D(1000))
```
