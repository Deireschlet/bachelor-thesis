```python
import numpy as np
import random
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
```

```python
def randomwalk1D(n):
    x = 0
    start = x
    xposition = [start]
    probabilities = [-1, 1]
    for i in range(1, n + 1):
        x += random.choice(probabilities)
        xposition.append(x)
    return pd.Series(xposition)
```

```python
random_data = randomwalk1D(500)
integrated_random_data = random_data - random_data.shift(1)
random_data
```

```python
random_data.shift(1)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(random_data)
plt.plot(integrated_random_data)
plt.xlim(0, len(integrated_random_data)) # Adjust the x-axis scale
plt.legend(['Random Walk', 'Integrated Random Walk'])
plt.title('Random Walk vs. Integrated Random Walk')
plt.savefig('random_walk.png')
```
